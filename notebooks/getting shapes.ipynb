```python
import requests  # to access the web API
from IPython.display import Image, display  # to display image data in this notebook
from pprint import pprint  # to print data structures in a nicer way
```

```python
# The prototype echoSMs web API URL
api_URL = 'https://echosms-data-store-app-ogogm.ondigitalocean.app'
# api_URL = 'http://127.0.0.1:8000'

# Get a list of all specimens in the datastore.
d = requests.get(api_URL + '/v2/specimens')
specimens = d.json()

# Print out how many specimens were provided
print(f'Obtained {len(specimens)} specimens')

# Print out the id and scientific name for all specimens
print('specimen_uuid,specimen_scientific_name,specimen_name')
for s in specimens:
    print(s['uuid'] + ',' + s.get('species', 'unknown') + ',' + s.get('specimen_name', 'unknown'))

# Print out the metadata for one of the specimens in the list. There is no shape data in this
index = min(17, len(specimens)-1)
pprint(specimens[index])
```

```python
# Get the shape data for one specimen
specimen_url = api_URL + '/v2/specimen/' + specimens[index]['uuid']
d = requests.get(specimen_url + '/data')

# Print out the specimen data. Metadata and shape data are provided. 
pprint(d.json())
```

```python
# Get a plot of the shape for one of the specimens
d = requests.get(specimen_url + '/image')

# A couple of hoops to jump through to show the returned image in a jupyter notebook
display(Image(d.content))
```
